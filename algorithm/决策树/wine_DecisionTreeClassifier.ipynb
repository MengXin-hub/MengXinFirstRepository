导入红酒数据集

```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
wine
```

```python
wine.data
```

```python
wine.target
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data), pd.DataFrame(wine.target)], axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

分训练集与测试集

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
Xtest.shape
```

```python
Xtrain.shape
```

建立模型

```python
# 三行代码:实例化，训练模型，使用接口导出结果
clf = tree.DecisionTreeClassifier(criterion='entropy', random_state=30, splitter='random')
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
```

```python
score
```

画出分类树

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']
class_name = ['琴酒', '雪莉', '贝尔摩德']
import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_name,  # 索引
                                class_names=class_name,  # 类别
                                filled=True,  # 颜色
                                rounded=True,  # 外框
                                out_file = None)
graph = graphviz.Source(dot_data)
```

```python
graph
```

```python
clf.feature_importances_  # 打印最重要属性
```

```python
[*zip(feature_name,clf.feature_importances_)]
```

```python
# 拟合程度
score_train = clf.score(Xtrain, Ytrain)
score_train
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                 ,random_state=30
                                 ,splitter="random"
                                 ,max_depth=5
                                 ,min_samples_leaf=10
                                 ,min_samples_split=10
                                 )
clf = clf.fit(Xtrain, Ytrain)
dot_data = tree.export_graphviz(clf
                               ,feature_names= feature_name
                               ,class_names=["琴酒","雪莉","贝尔摩德"]
                               ,filled=True
                               ,rounded=True
                               ,out_file=None
                               )  
graph = graphviz.Source(dot_data)
graph
```

```python
clf.score(Xtrain,Ytrain)
```

```python
clf.score(Xtest,Ytest)
```

```python
# 超参数曲线判断参数选取：超参数的学习曲线，是一条以超参数的取值为横坐标，模型的度量指标为纵坐标的曲
# 线，它是用来衡量不同超参数取值下模型的表现的线。
import matplotlib.pyplot as plt

test = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                     ,criterion="entropy"
                                     ,random_state=30
                                     ,splitter="random"
                                     )
    clf = clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    test.append(score)
plt.plot(range(1,11),test,color="red",label="max_depth")
plt.legend()
plt.show()
```

```python
#apply返回每个测试样本所在的叶子节点的索引
clf.apply(Xtest)
```

```python
#predict返回每个测试样本的分类/回归结果
clf.predict(Xtest)
```
